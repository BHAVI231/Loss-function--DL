```python
def mae(y_truth,y_predicted):
  total_error = 0
  for yt,yp in zip(y_truth,y_predicted):
    total_error +=abs(yt-yp) # Absolute error
  print("Total_error:" ,total_error)

  mae = total_error/len(y_truth)
  print("MAE:" ,mae)
  return mae
```

```python
mae(y_truth,y_predicted)
```

```python
def log_loss(y_truth, y_predicted):
    y_predicted_new = [max(i,epsilon) for i in y_predicted]
    y_predicted_new = [min(i,1-epsilon) for i in y_predicted_new]
    y_predicted_new = np.array(y_predicted_new)
    return -np.mean(y_truth*np.log(y_predicted_new)+(1-y_truth)*np.log(1-y_predicted_new))
```

```python
log_loss(y_truth, y_predicted)
```
